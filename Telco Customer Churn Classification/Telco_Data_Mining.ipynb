<b>What is customer churn and why is it so much important?</b>
   
Customer churn is one of the most important metrics in running a company and growing business to calculate its value on the market. This is the not happiest indicator, but it is a number which could tell You the truth about migration of Your clients. This is not possible to keep all of Your clients during a long time, but this factor can help You to reduce this negative trend as much as possible. It costs more to acquire new customers than it does to retain existing customers. In fact, an increase in customer retention of just $5\%$ can create at least a $25\%$ increase in profit. 

Most simply, customer churn is the percentage of customers that stopped using your company's product or service during a certain time frame. Calculations are not very complicated - You have to only divide number of lost customers during chosen time period by the number of customers you had at the beginning of that time period. For example, at the 1st of January Your company have $500$ customers, but in the end of the six-month period this number dropped to $420$. Your churn rate is $16\%$, because You lost $16\%$ ($80$ from $500$) of Your customers. 

However, you can calculate churn rate in whatever way is best for your company like the value of recurring business lost or the percentage of recurring value lost.

<b>Why do telecommunication companies analyze customer churn so deeply?</b>

<li><u>To boost their profits</u></li>
$\newline$
Precise and deeply analysis of reasons why Your customers decide to churn creates a clear way to better understand how to retain them. That appears an opportunity for addtional profit as we mentioned above. Sometimes these numbers are more impressive - according to some research returning customers will likely spend 67% more on your company's products and services. For many companies it is already convincing enough to start anaylizing customer churn.
$\newline$
<li><u>To create a better customer experience</u></li>
$\newline$
Effective churn analysis contributes to a deeper understanding why customers are leaving. Analyze how and when churn occurs in a customer's lifetime with your company, and use that data to develop a set of retention activities to create a more comfortable customer experience and fulfill customer needs much better. For example, when Your agreement with telecommunication company comes to end, You could be expecting a phone call from assistant with an attractive offer (discount for new mobile phone or a little bit less standing charge) to extend the contract. This way You can grow a community of Your most loyal customers which will share their positive experiences with benefiting Your service.
$\newline$
<li><u>To optimize products and services proactively</u></li>
$\newline$
These days on the market there are a lot of telecommunication companies so the customers have a widely range of offers to choose and locate their money. Customer churn analysis gives companies a quite accurate prediction about customer preferences: key attributes they are looking for in products/services (e.g. length of the agreement, state-of-the-art phone etc.), features they are dissatisfied with (e.g. highly paid roaming) or triggers that provide people to churn. Improving in this area could help optimize company offer to the most important (and profitable) customers.
