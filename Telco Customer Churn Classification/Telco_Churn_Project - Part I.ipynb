<b>What is customer churn and why is it so important?</b>
   
Customer churn is one of the most important metrics in running a company and growing business to calculate its value on the market. This isn't the happiest indicator, but it is a number which could tell You the truth about migration of Your clients. This is not possible to keep all of Your clients during a long period of time, but this factor can help You to reduce this negative trend as much as possible. It costs more to acquire new customers than it does to retain existing customers. In fact, an increase in customer retention of just $5\%$ can create at least a $25\%$ increase in profit. 

Simpler, customer churn is the percentage of customers that stopped using your company's product or service during a certain time frame. Calculations are not very complicated - You only have to divide number of lost customers during chosen time period by the number of customers you had at the beginning of that period. For example, at the 1st of January Your company have $500$ customers, but in the end of the six-month period this number dropped to $420$. Your churn rate is $16\%$, because You lost $16\%$ ($80$ from $500$) of Your customers. 

However, you can calculate churn rate in whatever way is best for your company like the value of recurring business lost or the percentage of recurring value lost.

<b>Why do telecommunication companies analyze customer churn so deeply?</b>

<li><u>To boost their profits</u></li>
$\newline$
Precise and deep analysis of reasons why Your customers decide to churn creates a clear way to better understand how to retain them. That appears an opportunity for addtional profit as we mentioned above. Sometimes these numbers are more impressive - according to some research returning customers will likely spend 67% more on your company's products and services. For many companies it is already convincing enough to start anaylizing customer churn.
$\newline$
<li><u>To create a better customer experience</u></li>
$\newline$
Effective churn analysis contributes to a deeper understanding why customers are leaving. Analyze how and when churn occurs in a customer's lifetime with your company, and use that data to develop a set of retention activities to create a more comfortable customer experience and fulfill customer needs much better. For example, when Your agreement with telecommunication company comes to end, You could be expecting a phone call from assistant with an attractive offer (discount for new mobile phone or a little bit less standing charge) to extend the contract. This way You can grow a community of Your most loyal customers which will share their positive experiences with benefiting Your service.
$\newline$
<li><u>To optimize products and services proactively</u></li>
$\newline$
These days on the market, there are a lot of telecommunication companies so the customers have a wide range of offers to choose and locate their money. Customer churn analysis gives companies a quite accurate prediction about customer preferences: key attributes they are looking for in products/services (e.g. length of the agreement, state-of-the-art phone etc.), features they are dissatisfied with (e.g. highly paid roaming) or triggers that provide people to churn. Improving in this area could help optimize company offer to the most important (and profitable) customers.

<a name = "4.2"><h3>4.2 Models evaluation and parameters optimazation</h3> </a>

Now we can begin evaluate classification models and optimaze their parameters to as precise as possible predict if customer is going to churn. Therefor, we compare with each other a few most frequent models like logstic regression or random tree classifier and at the end decide which one is perfect one to our task.

```python
# Function which shows results of the model
def model_results_plots(model, x_train, x_test, y_train):
    '''Function which returns statistics about classification model: 
    confusion matrix, accuracy of the model and graphs with true and 
    false labels and ROC curve course
    
    : param model : fitted model to dataset
    : type model : class sklearn.model
    
    : param x_train : part of dataset which served to training model
    : type x_train : numpy.array
    
    : param x_test : part of dataset on which accuracy is tested
    : type x_test : numpy.array
    
    :returns: ROC curve accuracy, classification report, confusion
    matrix and ROC curve plots'''
    
    # Making predictions on train and test data
    train_class_predictions = model.predict(x_train)
    test_class_predictions = model.predict(x_test)

    # Get the probabilities on train and test
    train_prob_predictions = model.predict_proba(x_train)[:,1]
    test_prob_predictions = model.predict_proba(x_test)[:,1]
    
    # Comparison of accuracy score and AUC score for train and test datasets
    print('Train dataset accuracy: ', accuracy_score(y_train, train_class_predictions))
    print('Train dataset AUC:', roc_auc_score(y_train, train_prob_predictions))
    print()
    print('Test dataset accuracy: ', accuracy_score(y_test, test_class_predictions))
    print('Test dataset AUC:', roc_auc_score(y_test, test_prob_predictions))
    print()

    # Get the classification report
    print('Classification report:')
    print(classification_report(y_test, test_class_predictions))
    print()

    # Statistics for classification problems (Precision, Recall and F1 score)
    F1_score = f1_score(y_test, test_class_predictions)
    precision = precision_score(y_test, test_class_predictions)
    recall = recall_score(y_test, test_class_predictions)

    # Graphs for confusion matrix and roc curve course
    sns.set_style('white')
    fig = plt.figure(figsize = (12, 6))
    
    # Confuson matrix
    ax1 = fig.add_subplot(121)
    
    y_pred = model.predict(x_test)
    
    #ConfusionMatrixDisplay(y_test,  y_pred, display_labels = ['0', '1'], ax = ax1, 
    #                                        cmap = 'flare', colorbar = False)#.plot(ax = ax1, colorbar = False)
    ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred), display_labels = ['0', '1']).plot(ax = ax1, colorbar = False, cmap = 'flare')
    ax1.set_xlabel('Predicted label', fontsize = 11)
    ax1.set_ylabel('True label', fontsize = 11)
    ax1.set_title('Confusion matrix for model', fontsize = 12)
    
    # ROC Curve
    sns.set_style('darkgrid')
    ax2 = fig.add_subplot(122)
    
    fpr, tpr, thresholds = roc_curve(y_test,  test_prob_predictions)
    
    ax2.plot([0, 1], [0, 1], linestyle = '--')
    ax2.plot(fpr, tpr, label = 'ROC curve')
    ax2.set_xlabel('False Positive Rate', fontsize = 11)
    ax2.set_ylabel('True Positive Rate', fontsize = 11)
    ax2.set_title('ROC curve course', fontsize = 12)
    ax2.legend(loc = 'lower right', fontsize = 10)
    
    fig.suptitle('Graphs for selected model', fontsize = 14, weight = 'bold')
    
    return model, precision, recall, F1_score, accuracy_score(y_test, test_class_predictions), roc_auc_score(y_test, test_prob_predictions)
```

<b>a) Logistic Regression</b>

<a href="https://www.ibm.com/topics/logistic-regression">Logistic regression</a> estimates the probability of an event occurring, such as voted or didn’t vote, based on a given dataset of independent variables. Since the outcome is a probability, the dependent variable is bounded between 0 and 1. The intuition behind Decision Trees is that you use the dataset features to create yes/no questions and continually split the dataset until you isolate all data points belonging to each class. <b>g) Bagging Classifier</b>

Bagging method forms a training algorithm which builds several instances of predictors on random subsets of training data. If the sampling is done with replacement then it is called Bagging. Here replacement means for a predictor an instance may get sampled multiple times while some are not sampled at all. This brings more randomness in the sampling process. Finally after the predictors are trained, prediction is made by aggregating the predictions of all predictors.

A Bagging classifier is an ensemble meta-estimator that fits base classifiers each on random subsets of the original dataset and then aggregate their individual predictions. In our case we use the most popular one in this technique - Decision Tree Classifier.

In our case choosing the right metric is not as clear as we thought about it earlier. The easiest way would be choice the mothid which gives us the highest accuracy on the test set, but it is not the best one. Also both precision and recall are not enough, because of the fact that because precision informs us to what extent we can trust our positive predictions in a given class, and recall informs us how many observations we lost for a given class.

However, in our case we more focus on the F1-score and ROC AUC Score metrics, which combine in some way the results between precision and recall and calculate the harmonic mean based on these two statistics to respect both. ROC AUC Score is often used to comparison the classification models, beacuse it the diagnostic capacity of the binary classifier system because its discrimination threshold varies.
